```python
import os
import pickle
import pandas as pd
from data_augmentation import create_augmented_images  # 拡張関数をインポート

# ファイルパスを指定
pkl_file_path = '../../Shared/morimoto/kvasir-seg/kvasir.pkl'
augmented_pkl_file_path = '../../Shared/morimoto/kvasir-seg/aug.pkl'

# .pklファイルを読み込む
with open(pkl_file_path, 'rb') as file:
    data = pickle.load(file)

# データの表示（オプション）
print(data)

# 拡張用のパラメータ設定
image_size = (320, 320)  # 画像のサイズを指定（適宜変更してください）
save_to_dir_images = '../../Shared/morimoto/kvasir-seg/aug_images'  # 拡張後の画像を保存するディレクトリ
save_to_dir_masks = '../../Shared/morimoto/kvasir-seg/aug_masks'  # 拡張後のマスクを保存するディレクトリ
num_aug = 20  # 1つの画像に対して生成する拡張画像の数

# 画像を拡張し、aug_train_dfを受け取る
aug_train_df = create_augmented_images(
    train_df=data,
    image_size=image_size,
    save_to_dir_images=save_to_dir_images,
    save_to_dir_masks=save_to_dir_masks,
    num_aug=num_aug
)

# 拡張されたデータフレームを .pklファイルとして保存
with open(augmented_pkl_file_path, 'wb') as file:
    pickle.dump(aug_train_df, file)

# 保存完了メッセージ
print(f"拡張されたデータは {augmented_pkl_file_path} に保存されました。")
```
